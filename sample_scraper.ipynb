```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

# Set up the Chrome webdriver (make sure you have Chrome and ChromeDriver installed)
driver = webdriver.Chrome()
```

```python
url = 'https://www.google.com/travel/hotels'
driver.get(url)
```

```python
search_bar = driver.find_element(By.CLASS_NAME, 'II2One')
search_bar.clear()
search_bar.send_keys('Kerala')

time.sleep(10)

# Create an ActionChains object and press the Enter key
actions = ActionChains(driver)
actions.send_keys(Keys.RETURN)
actions.perform()
```

```python
# Find all <c-wiz> elements with the specific jsrenderer attribute
c_wiz_elements = driver.find_elements(By.XPATH, "//c-wiz[@jsrenderer='hAbFdb']")

# Iterate through the elements and extract their text or other data
for c_wiz in c_wiz_elements:
    # Extract the text or other data from the c-wiz element
    element_text = c_wiz.text
    print(element_text)
```

```python
# for extracting the ratings

ele=driver.find_elements(By.CLASS_NAME,'lA0BZ')
print(len(ele))
for i in ele:
    print(i.text)
```

```python
# for the hotel names

ele = driver.find_elements(By.CSS_SELECTOR,'.BgYkof.ogfYpf')
print(len(ele))
for i in ele:
    print(i.text)
```

```python
# for extracting the features

ele = driver.find_elements(By.CSS_SELECTOR,'.bX73z')
# print(len(ele))
for i in ele:
    print(i.text)
```

```python
# for extracting prices

ele = driver.find_elements(By.CSS_SELECTOR,'.kixHKb.flySGb')
count = 0
for i in ele:
    if i.text != '':
        print(i.text)
        count += 1

print(count)
```

This is the class name of the parent div element

kCsInf


```python
div = driver.find_elements(By.CSS_SELECTOR, '.kCsInf')

for x in div:
    hotel_name = x.find_element(By.CSS_SELECTOR, '.BgYkof.ogfYpf').text
    
    # Extract the hotel rating, if it exists
    try:
        hotel_rating = x.find_element(By.CSS_SELECTOR, '.lA0BZ').text
    except:
        hotel_rating = 'NA'
    
    # Extract the hotel price, if it exists
    try:
        hotel_price = x.find_element(By.CSS_SELECTOR, '.kixHKb.flySGb').text
    except:
        hotel_price = 'NA'
    
    # Extract the hotel features
    features = x.find_elements(By.CSS_SELECTOR, '.bX73z')
    hotel_features = []
    
    for feat in features:
        hotel_features.append(feat.text)
    
    # Fill in "NA" for any missing features
    while len(hotel_features) < 9:
        hotel_features.append('NA')

    print(hotel_name)
    print(hotel_rating)
    print(hotel_price)
    print(hotel_features)
    print('-----------------------------------------------------------------------------')
```
